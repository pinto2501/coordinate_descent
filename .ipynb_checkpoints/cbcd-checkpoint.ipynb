```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

## Initialize

```python
n = 100
c = 0.3
I = np.identity(n)
matrix = c * np.ones((n, n))
path = None
L = np.tril(matrix, k=-1) # moves boundary one step below the diagonal
A_c = np.identity(n) + L + L.T
C = np.linalg.inv(I+L) @ L.T
S = np.diag([1/(n) for i in range(n)])
R = I - S @ np.linalg.inv(I) @ A_c
```

```python
epochs = 2000
x0 = np.ones((n,1))
x = x0.copy()
x_star = np.zeros((n,1))
ccd_errors = []

for l in range(epochs):
    x = C @ x
    ratio = np.linalg.norm(x - x_star) / np.linalg.norm(x0 - x_star)
    ccd_errors.append(np.log(ratio))

slope, intercept = np.polyfit(range(epochs), ccd_errors, 1)
print(-slope)
plt.plot(ccd_errors)
plt.title(f"CCD (n={n}, c={c})")
```

```python
epochs = 2000
x0 = np.ones((n,1))
x = x0.copy()
x_star = np.zeros((n,1))
rcd_errors = []

for l in range(epochs):
    x = R @ x
    ratio = np.linalg.norm(x - x_star) / np.linalg.norm(x0 - x_star)
    rcd_errors.append(np.log(ratio))

slope, intercept = np.polyfit(range(epochs), rcd_errors, 1)
print(-slope)
plt.plot(rcd_errors)
plt.title(f"ERCD (n={n}, c={c})")
```

```python
def build_coordinate_matrices(n, c):
    mats = []
    for i in range(n):
        M = np.eye(n)
        M[i, i] = 0.0
        M[i, :] = -c  
        M[i, i] = 0.0  
        mats.append(M)
    return mats

mats = build_coordinate_matrices(n, c)

# check if mats is correct

for mat in mats:
    mul_mat = np.matmul(mat, mat)
    if np.array_equal(mat @ mat, mat):
        pass
    else:
        print("Error")

    
epochs = 2000
x0 = np.ones((n, 1))
x = x0.copy()
x_star = np.zeros((n, 1))
prcd_errors = []

for l in range(epochs):
    i = random.randint(0, n - 1)
    x = mats[i] @ x
    ratio = np.linalg.norm(x - x_star) / np.linalg.norm(x0 - x_star)
    prcd_errors.append(np.log(ratio))


slope, intercept = np.polyfit(range(epochs), prcd_errors, 1)
print(-slope)
plt.plot(prcd_errors)
plt.title(f"PRCD (n={n}, c={c})")
plt.xlabel("Epoch")
plt.ylabel("Log ratio")
```
